Загрузка и фильтрация данных

```python
#Загружаем дела и приговоры по статье 126 УК РФ ("Похищение"), объединяем в один датарфейм
from pathlib import Path
import pandas as pd

file_path = Path("data/cases_126.csv")

try:

    cases_126 = pd.read_csv(file_path, encoding='utf-8')
except pd.errors.ParserError:
    try:
    
        cases_126 = pd.read_csv(file_path, encoding='utf-8', sep=';')
    except pd.errors.ParserError:
    
        cases_126 = pd.read_csv(file_path, encoding='utf-8', header=None)

file_path = Path("data/texts_126.csv")

try:

    text_126 = pd.read_csv(file_path, encoding='utf-8')
except pd.errors.ParserError:
    try:
    
        text_126 = pd.read_csv(file_path, encoding='utf-8', sep=';')
    except pd.errors.ParserError:
    
        text_126 = pd.read_csv(file_path, encoding='utf-8', header=None)

cases_126 = cases_126[cases_126['instance'] == 'FIRST']
cases_126 = cases_126[cases_126['decision'].str.contains("Вынесен ПРИГОВОР", case=False, na=False)]
merged_126 = cases_126.merge(text_126[['case_id', 'text']], on='case_id', how='left')
merged_126 = merged_126[~merged_126['text'].str.contains("постановление о прекращении уголовного дела", na=False)]
merged_126 = merged_126[merged_126["entry_date"] >= "2010-01-01"]
#merged_126 = merged_126.dropna(subset=['text'])
merged_126 = merged_126.drop_duplicates(subset='case_number', keep='first')


merged_126 = merged_126[~((merged_126['accused'].str.contains('127\.1|127\.2', na=False)) | 
                   (merged_126['text'].str.contains('127\.1|127\.2', na=False)))]

merged_126.reset_index(drop=True, inplace=True)
merged_126.head(5)
```

```python
merged_126.to_csv('dataset_126.csv', index=False)
```

Оцениваем процент закрытых приговоров по каждой статье

```python
import pandas as pd


merged_126['year'] = pd.to_datetime(merged_126['entry_date']).dt.year

#Cоздадим колонку is_closed — 1 если текст приговора закрытый, 0 если открытый
merged_126['is_closed'] = merged_126['text'].isna().astype(int)

#По регионам
region_stats = merged_126.groupby('region_name')['is_closed'].mean() * 100
region_mean = region_stats.mean()
region_std = region_stats.std()

print(f'Средний процент закрытых по регионам: {region_mean:.2f}%')
print(f'Стандартное отклонение по регионам: {region_std:.2f}%')

#По годам 
year_stats = merged_126.groupby('year')['is_closed'].mean() * 100
year_mean = year_stats.mean()
year_std = year_stats.std()

print(f'\nСредний процент закрытых по годам: {year_mean:.2f}%')
print(f'Стандартное отклонение по годам: {year_std:.2f}%')
```
